## Production

```python
import statsmodels.api as sm
from sklearn import linear_model
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Cleaning the Data Set Further for Regression
top_pr = pd.read_csv(r"C:\Users\Seagan\Downloads\DLSU\DLSU STUDY\Year 3\DATA100\transposed_top_production.csv")
top_pr = top_pr.drop(labels=0, axis=0).reset_index(drop=True)
top_pr["Brazil"] = top_pr["Brazil"].astype(float)
top_pr["Indonesia"] = top_pr["Indonesia"].astype(float)
top_pr["Colombia"] = top_pr["Colombia"].astype(float)
top_pr["Ethiopia"] = top_pr["Ethiopia"].astype(float)
top_pr["Viet Nam"] = top_pr["Viet Nam"].astype(float)
top_pr[["Brazil", "Indonesia", "Colombia", "Ethiopia", "Viet Nam"]] = top_pr[["Brazil", 
 "Indonesia", "Colombia", "Ethiopia", "Viet Nam"]].astype(np.int64)
top_pr = top_pr.drop(labels=30, axis=0)
top_pr = top_pr.rename(columns={top_pr.columns[0]: "Year"})
top_pr['Year'] = top_pr['Year'].str[:4]
print(top_pr)
```

```python
#Regressing Brazil Production on Year
x = top_pr["Year"].astype(np.int64)
y = top_pr["Brazil"]
x = sm.add_constant(x)

regr = linear_model.LinearRegression()
regr.fit(x,y)

model = sm.OLS(y,x).fit()
predictions = model.predict(x)
print_model = model.summary()
print(print_model)
```
